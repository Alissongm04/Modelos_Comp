```python
import networkx as nx
import matplotlib.pyplot as plt

#creamos un grafo vacío 
G = nx.DiGraph()

#Agregamos nodos al grafico 
G.add_nodes_from(['A', 'B', 'C', 'D', 'E'])

G.nodes
```

```python
nx.draw_networkx(G) #aquí no tienen conexión ni dirección
```

```python

G.add_edges_from([('A', 'B', {'weight':2.5}), 
                  ('B', 'C', {'weight':1.5}), 
                  ('C', 'A', {'weight':2.0}), 
                  ('D', 'E', {'weight':1.0})])

G.edges
```

```python
G.adj
```

```python
list(G.adjacency())
```

```python
pos = nx.spring_layout(G)
pos
```

```python
nx.draw_networkx(G, pos, arrows=True, arrowstyle='-|>')
plt.grid(True)
plt.axis('on')
plt.show()
```

```python
# nodes
nx.draw_networkx_nodes(G, pos, node_size=400)
# edges
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=4)
# labels
# node labels
nx.draw_networkx_labels(G, pos, font_size=20)
# edge weight labels
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels)
```

Eliminando aristas

```python
G.remove_edge('D', 'E')
nx.draw_networkx(G, with_labels=True, arrows=True)
```

Grafo a dataframe

```python
nx.to_pandas_adjacency(G)

nx.to_pandas_edgelist(G)
```
