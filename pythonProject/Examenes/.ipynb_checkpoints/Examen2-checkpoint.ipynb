1. Crear una función llamada "df_ordered" que reciba un dataframe y un str y retorne un dataframe ordenado de mayor a menor considerando la característica str

```python
data = df_prevalencia_global
def df_ordered(df, column):
    return df.sort_values(by=column, ascending=False)

df = pd.DataFrame(data)

df_ordered(df, 'Number of countries with primary data on prevalence of mental disorders')
```

2. Crear una función llamada "plot_bar" que reciba un dataframe y un str y haga el plot de barras de ese dataframe considerando esa característica descrita por el string

```python
data2 = df_ansiedad
def plot_bar(df, column):
    return df.plot(kind='bar', stacked=True)

df2 = pd.DataFrame(data2)

plot_bar(df2, 'year')
```

3. Ordenar países con mayor depresión


```python
data = df_prevalencia_depresion
def df_ordered(df, column):
    return df.sort_values(by=column, ascending=False)

df = pd.DataFrame(data)

df_ordered(df, 'Major depression')
```

4. Realiza una gráfica de barras mostrando el valor de depresión y la entidad (país) 

```python
import numpy as np 
import matplotlib.pyplot as plt 
data4 = df_prevalencia_depresion

def plot_multiple_bar(df,category_column, value_columns):
    index = np.arange(len(df4[category_column]))
    
    
    for i, col in enumerate(value_columns):
        plt.bar(index + i, df4[col], label=col)
        
    plt.xticks(index, df[category_column])
    
    plt.figure(figsize=(20,4))
    plt.xlabel(fontsize=7)
    plt.ylabel(fontsize=14)
    

plot_multiple_bar(data4, 'Entity', ['Major depression'])
```
